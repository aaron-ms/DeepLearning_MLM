# Save a Keras Model

```python
# MLP for pima Indians Dataset saved to single file
from numpy import loadtxt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input
from tensorflow.keras.layers import Dense

# load pima indians datasets
dataset = loadtxt("datasets/pima-indians-diabetes.csv", delimiter=',')

# split into input (X) and output (Y) variables
X = dataset[:, 0:8]
Y = dataset[:, 8]

# define the Model
model = Sequential()
model.add(Input(shape=(8, )))
model.add(Dense(12, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

# Compile the model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Fit the model
model.fit(X, Y, epochs=150, batch_size=10, verbose=1)

# Evaluate the model
scores = model.evaluate(X, Y, verbose=0)
print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))

# save the model and architecture in a single file
model.save('model.keras')
print('Save model to disk.')
```

```python
# another way to save the model 
# .h5 format is legacy hence we need to save .keras format
from tensorflow.keras.models import save_model

save_model(model, 'model.keras')
```

```python
# Load a Keras Model
from numpy import loadtxt
from tensorflow.keras.models import load_model

# load Model
model = load_model('model.keras')

# summarizes model.
model.summary()

# load dataset
dataset = loadtxt('datasets/pima-indians-diabetes.csv', delimiter=',')

# split into input (X) and Output (Y) variable
X = dataset[:, 0:8]
Y = dataset[:, 8]

# evaluate the model
score = model.evaluate(X, Y, verbose=0)
print("%s: %.2f%%" % (model.metrics_names[1], score[1]*100))
```

# Protocol Buffer Format
While saving and loading a Keras model using HDF5 format is the recommended way, TensorFlow
supports yet another format, the protocol buffer. It is considered faster to save and load a protocol buffer
format, but doing so will produce multiple files. The syntax is the same, except that you do not need to
provide the .h5/.keras extension to the filename:

```python
# Protocol Buffer Format
# MLP for pima Indians Dataset saved to single file
from numpy import loadtxt
from tensorflow.keras.models import Sequential
from tensorflow.keras.models import load_model
from tensorflow.keras.layers import Input
from tensorflow.keras.layers import Dense

# load pima indians datasets
dataset = loadtxt("datasets/pima-indians-diabetes.csv", delimiter=',')

# split into input (X) and output (Y) variables
X = dataset[:, 0:8]
Y = dataset[:, 8]

# define the Model
model = Sequential()
model.add(Input(shape=(8, )))
model.add(Dense(12, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

# Compile the model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Fit the model
model.fit(X, Y, epochs=150, batch_size=10, verbose=1)

# Evaluate the model
scores = model.evaluate(X, Y, verbose=0)
print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))

# save the model and architecture in a single file
model.save('model')
print('Save model to disk.')

# load model
model = load_model('model')

# summary
model.summary()
```
